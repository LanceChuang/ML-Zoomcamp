```python
df_innjoo = dataset[['RAM', 'Storage', 'Screen']]
X = df_innjoo.values
# Compute XTX
XTX = np.dot(X.T, X)
# Compute the inverse of XTX
XTX_inv = np.linalg.inv(XTX)
# Create the array y
y = np.array([1100, 1300, 800, 900, 1000, 1100])
w = np.dot(np.dot(XTX_inv, X.T ), y)
# Calculate the sum of elements in w
sum_w = np.sum(w)

print("Sum of w:", sum_w)
```
